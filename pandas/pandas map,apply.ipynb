```python
import pandas as pd
df=pd.DataFrame({'name':['anila','henry','caven'],
                'gender':[1,2,0],
                'age':[80,38,12],
                'height':[147.2,173.5,180.5],
                'weight':[43,45,85]
                
                })


gender_map={0:'unknown', 1:'male',2:'female'}
df['gender']=df['gender'].map(gender_map)
df
```

```python
# apply function
import pandas as pd
df=pd.DataFrame({'name':['anila','henry','caven'],
                'gender':[1,2,0],
                'age':[80,38,12],
                'height':[147.2,173.5,180.5],
                'weight':[43,45,85]
                
                })

import numpy as np
df['height']=df['height'].apply(np.ceil)
df

```
