```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,16])
plt.bar(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["D","F","G","I"])
y=np.array([5,8,7,10])
plt.barh(x,y,color="black")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","N","I","L","A"])
y=np.array([12,21,20,18,14])
plt.bar(x,y, color ="blue")
plt.show()
```

```python
#        HISTOGRAM
# ------------------------------

import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```
